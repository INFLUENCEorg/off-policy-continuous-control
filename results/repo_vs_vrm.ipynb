```python
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats
%matplotlib inline
set_matplotlib_formats('svg')
img = plt.imread("vrm/cartpole.png")
from plot_utils import eleven_neighbor_smooth
from plot_avg_return import plot_all_runs
```

```python
import pandas as pd
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(5, 5))

estimated_low = 0
estimated_high = 1050
num_epoch = 300
num_timesteps = int(3e5)

ax.imshow(img, interpolation='none', extent=[0, num_timesteps, estimated_low, estimated_high])
ax.set_aspect(num_timesteps/(estimated_high - estimated_low))

xs = np.linspace(0, num_timesteps, num_epoch)

plot_all_runs('cartpole-continuous-long-v0', xs=xs)

plt.ylim(estimated_low, estimated_high)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))

plt.legend(loc='lower right')
plt.xlabel('No. of env interactions')
plt.ylabel('Test-time Return')
plt.title('Learning curve for cartpole-continuous-long-v0\n(purple is SAC from VRM paper)')
plt.show()
```
